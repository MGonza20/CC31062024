```python
import math
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# 0. Cargar los datos de estatura y peso
url = 'http://wiki.stat.ucla.edu/socr/index.php?title=SOCR_Data_Dinov_020108_HeightsWeights&oldid=12529'

height_weight_df = pd.read_html(url)[1][['Height(Inches)','Weight(Pounds)']]
```

```python
height_weight_df
```

```python
# 1. Count records in the dataframe
# 2. Crear una regresion lineal con estatura como variable explicatoria y peso como variable explicada
# 3. Hallar el intercepto
# 4. Hacer la grafica
# 5. Hallar correlacion y valor p
# 6. Interpretar el valor p
```

```python
# 1. Count records in the dataframe
print(f'Records count: {len(height_weight_df)}')
```

```python
# 2. Crear una regresion lineal con estatura como variable explicatoria y peso como variable explicada

X = height_weight_df[['Height(Inches)']] # Variable explicatoria
y = height_weight_df['Weight(Pounds)'] # Variable explicada

lm = linear_model.LinearRegression()
lm.fit(X, y)
```

```python
# 3. Hallar el intercepto
intercepto = lm.intercept_
print(f'Intercepto: {intercepto}')
```

```python
# 4. Hacer la grafica
plt.scatter(X, y)
plt.plot(X, lm.predict(X), color='orange')
plt.xlabel('Estatura (Variable explicatoria)')
plt.ylabel('Peso (Variable explicada)')
plt.show()
```

```python
# 5. Hallar correlacion y valor p
corr, val_p = pearsonr(height_weight_df['Height(Inches)'], height_weight_df['Weight(Pounds)'])
print(f'CorrelaciÃ³n: {corr}')
print(f'Valor p: {val_p}')
```

### 6. Interpretar el valor p

El valor p es menor a 0.05, lo cual permite rechazar la hipÃ³tesis nula y concluir que existe una correlaciÃ³n entre la estatura y el peso. Debido a que la correlaciÃ³n es positiva, es posible afirmar que a mayor estatura, mayor es el peso. Este valor p indica que los resultados observados no son aleatorios, sino que reflejan un efecto real.
